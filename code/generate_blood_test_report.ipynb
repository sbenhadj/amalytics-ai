```python
from lxml import etree
from collections import defaultdict

# Load your XML file
tree = etree.parse("report.xml")  # replace with your actual path
root = tree.getroot()

# Group Param elements by parent path and name
param_groups = defaultdict(list)
for param in root.xpath(".//Param"):
    name = param.get("name")
    unit = param.get("unit", "")
    parent_path = tree.getpath(param.getparent())
    key = (parent_path, name)
    param_groups[key].append(param)

# Merge entries with same name but different units
for (path, name), params in param_groups.items():
    if len(params) > 1:
        main_param = params[0]
        alt_units = []
        for p in params[1:]:
            u = p.get("unit")
            if u and u != main_param.get("unit"):
                alt_units.append(u)
            p.getparent().remove(p)
        if alt_units:
            main_param.set("alt_unit", ", ".join(alt_units))

# Save the modified XML
tree.write("modified_report.xml", encoding="UTF-8", pretty_print=True, xml_declaration=True)
```

```python
from lxml import etree
from collections import defaultdict
from itertools import combinations

# Load your XML
tree = etree.parse("report.xml")
root = tree.getroot()

# Map each parameter name to all its observed units
unit_map = defaultdict(set)
for param in root.xpath(".//Param"):
    name = param.get("name")
    unit = param.get("unit")
    if name and unit:
        unit_map[name].add(unit)

# Find all unique unit conversion pairs across parameters
conversion_pairs = set()
for units in unit_map.values():
    if len(units) > 1:
        for a, b in combinations(units, 2):
            conversion_pairs.add(tuple(sorted((a, b))))

# Display
print("Possible unit conversion pairs:")
for pair in sorted(conversion_pairs):
    print(f"{pair[0]} <--> {pair[1]}")
```

```python
from fpdf import FPDF
import random
import datetime

def generate_patient_info(prenoms_hommes, prenoms_femmes, noms_famille, medecins, biologistes):
    # Random sex (M or F)
    sexe = random.choice(["M", "F"])

    if sexe == "M":
        prenom = random.choice(prenoms_hommes)
        civilite = "M."   # Monsieur
    else:
        prenom = random.choice(prenoms_femmes)
        civilite = "Mme"  # Madame
    nom_famille = random.choice(noms_famille)

    age = random.randint(18, 90)  # Random age between 18 and 90
    annee_naiss = datetime.date.today().year - age
    mois_naiss = random.randint(1, 12)
    jour_naiss = random.randint(1, 28)  # Simplification: day 1-28
    date_naissance = datetime.date(annee_naiss, mois_naiss, jour_naiss)

    # Generate a unique dossier number (e.g., date + random digits)
    dossier_num = f"{datetime.date.today():%d%m%y}-" + f"{random.randint(1000,9999)}"

    # Date and time of collection and report creation
    date_prelev = datetime.datetime.now() - datetime.timedelta(hours=random.randint(1, 72))
    date_edit = datetime.datetime.now()

    # Prescribing doctor and validating biologist
    medecin_prescripteur = random.choice(medecins)
    biologiste_validateur = random.choice(biologistes)

    # Gather patient information
    patient_info = {
        "civilite": civilite,
        "nom": nom_famille.upper(),       # Name in uppercase by convention
        "prenom": prenom,
        "sexe": "F" if sexe == "F" else "M",
        "date_naiss": date_naissance.strftime("%d/%m/%Y"),
        "age": age,
        "adresse": "123 Rue Exemple, 75000 PARIS",
        "dossier": dossier_num,
        "date_prelev": date_prelev.strftime("%d/%m/%y %Hh%M"),
        "date_edit": date_edit.strftime("%d/%m/%y à %Hh%M"),
        "medecin": medecin_prescripteur,
        "biologiste": biologiste_validateur
    }

    return patient_info
```

```python
import os
import json
import random
from fpdf import FPDF
from fpdf.enums import XPos, YPos

pdf_output_dir = "reports"
json_output_dir = "extracted"

os.makedirs(pdf_output_dir, exist_ok=True)
os.makedirs(json_output_dir, exist_ok=True)

num_samples = 1000

for i in range(num_samples):
    resultats = {}
    filled_template = {}

    for categorie in root:
        cat_name = categorie.tag
        resultats[cat_name] = {}
        filled_template[cat_name] = {}

        for sous_categorie in categorie:
            sous_cat_name = sous_categorie.tag
            resultats[cat_name][sous_cat_name] = []
            filled_template[cat_name][sous_cat_name] = {}

            selection_ratio = random.choices([0.0, 0.3, 0.5, 0.7, 0.8, 0.9, 1.0], weights=[0.5, 0.2, 0.3, 0.1, 0.2, 0.3, 0.8])[0]

            all_params = sous_categorie.findall('Param')
            num_to_select = int(len(all_params) * selection_ratio)
            mask = [True] * num_to_select + [False] * (len(all_params) - num_to_select)
            random.shuffle(mask)

            for param in all_params:
                nom = param.get('name')
                unite = param.get('unit')
                if unite == "qualitative":
                    filled_template[cat_name][sous_cat_name][nom] = {
                        "valeur": None,
                    }
                else:
                    filled_template[cat_name][sous_cat_name][nom] = {
                        "valeur": None,
                        "unité": unite
                    }

            # Select parameters based on mask
            selected_params = [param for param, keep in zip(all_params, mask) if keep]

            #Enforce exclusivity between "Globules rouges" and "Hématies"
            names_in_selection = [p.get('name') for p in selected_params]
            if "Globules rouges" in names_in_selection and "Hematies" in names_in_selection:
                keep_name = random.choice(["Globules rouges", "Hematies"])
                remove_name = "Hematies" if keep_name == "Globules rouges" else "Globules rouges"

                # Remove the unwanted one from selected params
                selected_params = [p for p in selected_params if p.get('name') != remove_name]

                # Also remove from filled_template
                if remove_name in filled_template[cat_name][sous_cat_name]:
                    del filled_template[cat_name][sous_cat_name][remove_name]


            # Special values
            HDL = LDL = triglycerides = None
            cholesterol_insert_index = None
            Bconjuguée1 = Blibre1 = None
            b_insert_index = None
            val_testo = None

            for idx, param in enumerate(selected_params):
                nom = param.get('name')
                unit = param.get('unit').replace("μ", "u") or ""
                normal_min = param.get('normal_min')
                normal_max = param.get('normal_max')
                pathological_low = param.get('pathological_low')
                pathological_high = param.get('pathological_high')
                alt_unit = param.get('alt_unit')

                if unit == "qualitative":
                    values_str = param.get('values')
                    values = values_str.split(',') if values_str else []
                    valeur = random.choice(values) if values else None
                    filled_template[cat_name][sous_cat_name][nom] = {"valeur": valeur}
                    resultats[cat_name][sous_cat_name].append({
                        'name': nom,
                        'value': valeur,
                    })
                else:
                    valeur = None
                    valeur_alt = None

                    if normal_min is None and normal_max is None:
                        valeur = pathological_high if pathological_high and random.random() < 0.1 else param.get('normal') or ""
                    else:
                        low = float(normal_min) if normal_min else 0.0
                        high = float(normal_max) if normal_max else float(normal_min)
                        etat = random.choices(["normal", "low", "high"], weights=[0.7, 0.15, 0.15])[0]
                        if etat == "normal":
                            valeur = random.uniform(low, high)
                        elif etat == "low":
                            valeur = low * (1 - random.uniform(0.1, 0.3))
                        elif etat == "high":
                            valeur = high * (1 + random.uniform(0.1, 0.3))

                        if unit in ["T/L", "G/L", "mmol/L"]:
                            valeur = round(valeur, 2)
                        elif unit in ["%", "pg", "fL", "g/dL"]:
                            valeur = round(valeur, 1)
                        else:
                            valeur = round(valeur, 1)

                    # Capture specific values for synthetic computation
                    if nom == "Cholestérol H.D.L." and unit == "mmol/L": HDL = float(valeur)
                    if nom == "Cholestérol L.D.L." and unit == "mmol/L": LDL = float(valeur)
                    if nom == "Triglycérides" and unit == "mmol/L": triglycerides = float(valeur)
                    if nom == "Cholestérol non-HDL": cholesterol_insert_index = len(resultats[cat_name][sous_cat_name])
                    if nom == "Bilirubine conjuguée" and unit == "umol/L": Bconjuguée1 = float(valeur)
                    if nom == "Bilirubine libre" and unit == "umol/L":
                        Blibre1 = float(valeur)
                        b_insert_index = len(resultats[cat_name][sous_cat_name])
                    if nom == "Testostérone": val_testo = valeur

                    # Default to main value and unit
                    valeur_for_json = valeur
                    unit_for_json = unit

                    # 20% chance to select alternative unit if available and conversion possible
                    if alt_unit and nom in conversion_factors and isinstance(valeur, (int, float)) and random.random() < 0.2:
                        try:
                            valeur_alt =  round(convert_main_to_alt(valeur, nom),2)
                            if valeur_alt is not None:
                                valeur_for_json = valeur_alt
                                unit_for_json = alt_unit
                        except Exception as e:
                            print(f"[WARN] Conversion error for {nom}: {e}")

                    filled_template[cat_name][sous_cat_name][nom] = {
                        "valeur": valeur_for_json,
                        "unité": unit_for_json
                    }


                    resultat_str = f"{valeur} {unit}".strip()
                    if valeur_alt is not None:
                        alt_unit_short = alt_unit.split()[0]
                        resultat_str += f" ({valeur_alt} {alt_unit_short})"

                    resultats[cat_name][sous_cat_name].append({
                        'name': nom,
                        'value': resultat_str,
                        'normal_min': normal_min,
                        'normal_max': normal_max
                    })
            if sous_cat_name == "FormuleLeucocytaire":
            # Generate Leucocytes value
              leucocytes_val = round(random.uniform(4000, 11000), 0)  # Example range /mm3
              filled_template[cat_name][sous_cat_name]["Leucocytes"] = {
                  "valeur": leucocytes_val,
                  "unité": "/mm3"
              }
              resultats[cat_name][sous_cat_name].append({
                  'name': "Leucocytes",
                  'value': f"{leucocytes_val} /mm3",
              })

              # 2) Derivatives and percentages
              derivatives = ["Neutrophiles", "Eosinophiles", "Basophiles", "Lymphocytes", "Monocytes"]
              # Generate random percentages summing to 100
              random_percents = [random.uniform(1, 20) for _ in derivatives]  # positive weights
              total = sum(random_percents)
              percentages = [round(p/total*100) for p in random_percents]

              # Adjust sum to exactly 100%
              diff = 100 - sum(percentages)
              percentages[0] += diff  # fix rounding error on first element

              for deriv, perc in zip(derivatives, percentages):
                  # Fill template with only percentage
                  filled_template[cat_name][sous_cat_name][deriv] = {
                      "valeur": perc,
                      "unité": "%"
                  }

                  # PDF value = "XX% : YY G/L"

                  absolute_val = round(leucocytes_val * perc / 100, 2)
                  pdf_val = f"{perc}% : {absolute_val} /mm3"

                  resultats[cat_name][sous_cat_name].append({
                      'name': deriv,
                      'value': pdf_val,
                  })

              continue

            # Cholesterol logic
            if cat_name == "Biochimie" and sous_cat_name == "Biochimie":
                if HDL is not None and LDL is not None and triglycerides is not None:
                    chol_total1 = round(HDL + LDL + (triglycerides / 2.2), 2)
                    chol_total2 = round(chol_total1 * 0.386, 2)

                    filled_template[cat_name][sous_cat_name]["Cholestérol total"] = {
                        "valeur": chol_total1,
                        "unité": "mmol/L",
                    }

                    chol_result = {
                        "name": "Cholestérol total",
                        "value": f"{chol_total1} mmol/L ({chol_total2} g/L)",
                        "normal_min": None,
                        "normal_max": "5.18"
                    }

                    if cholesterol_insert_index is not None:
                        resultats[cat_name][sous_cat_name].insert(cholesterol_insert_index, chol_result)
                    else:
                        resultats[cat_name][sous_cat_name].append(chol_result)
                else:
                    filled_template[cat_name][sous_cat_name]["Cholestérol total"] = {"valeur": None, "unité": "mmol/L"}

                # Reorder
                ft = filled_template[cat_name][sous_cat_name]
                if "Cholestérol total" in ft:
                    reordered = {}
                    for key in ft:
                        reordered[key] = ft[key]
                        if key == "Triglycérides":
                            reordered["Cholestérol total"] = ft["Cholestérol total"]
                    filled_template[cat_name][sous_cat_name] = reordered

                # Bilirubine totale
                if Bconjuguée1 is not None and Blibre1 is not None:
                    Btotal1 = round(Bconjuguée1 + Blibre1, 2)
                    Btotal2 = round(Btotal1 * 0.5847, 2)

                    filled_template[cat_name][sous_cat_name]["Bilirubine totale"] = {
                        "valeur": Btotal1,
                        "unité": "umol/L",
                    }

                    B_result = {
                        "name": "Bilirubine totale",
                        "value": f"{Btotal1} umol/L ({Btotal2} mg/L)",
                        "normal_min": "3.4",
                        "normal_max": "20.5"
                    }

                    if b_insert_index is not None:
                        resultats[cat_name][sous_cat_name].insert(b_insert_index, B_result)
                    else:
                        resultats[cat_name][sous_cat_name].append(B_result)
                else:
                    filled_template[cat_name][sous_cat_name]["Bilirubine totale"] = {"valeur": None, "unité": "µmol/L"}

                ft = filled_template[cat_name][sous_cat_name]
                if "Bilirubine totale" in ft:
                    reordered = {}
                    for key in ft:
                        reordered[key] = ft[key]
                        if key == "Bilirubine libre":
                            reordered["Bilirubine totale"] = ft["Bilirubine totale"]
                    filled_template[cat_name][sous_cat_name] = reordered

            # Testosterone fix
            if cat_name == "Hormonologie" and sous_cat_name == "Hormonologie":
                if val_testo is not None:
                    filled_template[cat_name][sous_cat_name]["Testostérone"] = {
                        "valeur": f"Inf à {val_testo}",
                        "unité": "ng/ml",
                    }

                    if resultats[cat_name][sous_cat_name]:
                        resultats[cat_name][sous_cat_name][-1]["value"] = f"Inf à {val_testo} ng/ml"
                    else:
                        resultats[cat_name][sous_cat_name].append({
                            "name": "Testostérone",
                            "value": f"Inf à {val_testo} ng/ml",
                            "normal_min": None,
                            "normal_max": None
                        })

    # Output generation
    pdf_filename = f"{pdf_output_dir}/sample_{i+1}.pdf"
    generate_pdf(resultats, pdf_filename)

    json_filename = f"{json_output_dir}/sample_{i+1}_template.json"
    with open(json_filename, 'w', encoding="utf-8") as json_file:
        json.dump(filled_template, json_file, ensure_ascii=False, indent=2)
```

```python
import json
import random
from collections import OrderedDict

def filter_with_custom_ratios(template):
    keep_ratios = {
        "NumerationGlobulaire": 0.6,
        "FormuleLeucocytaire": 1.0,
        "NumerationPlaquettaire": 1.0,
        "Hemostase": 0.6,
        "Biochimie": 0.4,
        "CytologieUrinaire": 0.5,
        "Hormonologie": 0.8
    }

    cholesterol_keywords = ["Cholestérol", "cholestérol"]
    bilirubine_fields = ["Bilirubine totale", "Bilirubine libre", "Bilirubine conjuguée"]

    filled_output = OrderedDict()
    empty_output = OrderedDict()

    for cat, subcats in template.items():
        filled_output[cat] = OrderedDict()
        empty_output[cat] = OrderedDict()

        for subcat, params in subcats.items():
            param_items = list(params.items())
            total = len(param_items)

            # --- Detect group triggers ---
            cholesterol_trigger = any(
                any(kw in name for kw in cholesterol_keywords) and val.get("valeur") not in [None, "", "null"]
                for name, val in param_items
            )

            bilirubine_trigger = any(
                name in bilirubine_fields and val.get("valeur") not in [None, "", "null"]
                for name, val in param_items
            )

            # --- Apply logic ---
            kept_fields = OrderedDict()
            kept_empty_fields = OrderedDict()

            if cholesterol_trigger or bilirubine_trigger:
                for name, val in param_items:
                    if cholesterol_trigger and any(kw in name for kw in cholesterol_keywords):
                        # Keep cholesterol fields
                        kept_fields[name] = val
                        empty_val = {"valeur": None}
                        if "unité" in val:
                            empty_val["unité"] = val["unité"]  # keep original unit
                        kept_empty_fields[name] = empty_val
                    elif bilirubine_trigger and name in bilirubine_fields:
                        # Keep bilirubine fields
                        kept_fields[name] = val
                        empty_val = {"valeur": None}
                        if "unité" in val:
                            empty_val["unité"] = val["unité"]  # keep original unit
                        kept_empty_fields[name] = empty_val
                    else:
                        # Possibly keep other fields based on ratio
                        pass  # do nothing now
            else:
                # Default random selection by keep_ratio
                ratio = keep_ratios.get(subcat, 1.0)
                num_to_keep = max(1, int(total * ratio)) if total > 0 else 0
                keep_indices = set(random.sample(range(total), num_to_keep))

                for i, (name, val) in enumerate(param_items):
                    if i in keep_indices:
                        kept_fields[name] = val
                        empty_val = {"valeur": None}
                        if "unité" in val:
                            empty_val["unité"] = val["unité"]  # keep original unit
                        kept_empty_fields[name] = empty_val

            # Combine any leftover logic (keep all other non-triggered fields if selected by ratio)
            if cholesterol_trigger or bilirubine_trigger:
                # Apply default ratio to the remaining fields not part of the trigger group
                other_items = [
                    (i, (name, val)) for i, (name, val) in enumerate(param_items)
                    if not (cholesterol_trigger and any(kw in name for kw in cholesterol_keywords))
                    and not (bilirubine_trigger and name in bilirubine_fields)
                ]
                ratio = keep_ratios.get(subcat, 1.0)
                num_to_keep = max(1, int(len(other_items) * ratio)) if other_items else 0
                keep_indices = set(random.sample(range(len(other_items)), num_to_keep))

                for idx, (i, (name, val)) in enumerate(other_items):
                    if idx in keep_indices:
                        kept_fields[name] = val
                        empty_val = {"valeur": None}
                        if "unité" in val:
                            empty_val["unité"] = val["unité"]  # keep original unit
                        kept_empty_fields[name] = empty_val

            filled_output[cat][subcat] = kept_fields
            empty_output[cat][subcat] = kept_empty_fields

    return filled_output, empty_output
```

```python
input_folder = "/content/drive/MyDrive/ground_truth/templates"

# Output folders for filled and empty templates
filled_folder = "/content/filtered_templates/filled"
empty_folder = "/content/filtered_templates/empty"

# Create output directories if they don't exist
os.makedirs(filled_folder, exist_ok=True)
os.makedirs(empty_folder, exist_ok=True)

# Process each JSON file in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith(".json"):
        file_path = os.path.join(input_folder, filename)

        # Load the template
        with open(file_path, "r", encoding="utf-8") as f:
            original_template = json.load(f, object_pairs_hook=OrderedDict)

        # Apply filtering
        filled, empty = filter_with_custom_ratios(original_template)

        base_name = os.path.splitext(filename)[0]  # e.g., "sample_1_template"

        # Save filled template
        filled_path = os.path.join(filled_folder, f"{base_name}_filled.json")
        with open(filled_path, "w", encoding="utf-8") as f:
            json.dump(filled, f, indent=2, ensure_ascii=False)

        # Save empty template
        empty_path = os.path.join(empty_folder, f"{base_name}_empty.json")
        with open(empty_path, "w", encoding="utf-8") as f:
            json.dump(empty, f, indent=2, ensure_ascii=False)

        print(f"Processed {filename} → filled + empty saved.")

```
